```python
#Correlation Matrix
# Compute the pair wise corelation for all columns
correlation_matrix = data.corr().round(2)
correlation_matrix
```

```python
#use the heatmap function from seaborn to plot the correlation matrix
#Plot rectangular data as a color-encoded matrix.
# annot = True to print the values inside the square
sbn.set(rc = {"figure.figsize":(12,10)})
sbn.heatmap(data = correlation_matrix , annot = True)
```

```python
plt.figure(figsize=(20, 5))

features = ['RM', 'DIS']
target = data['MEDV']

#Enumerate() method adds a counter to an iterable and returns it in a form of enumerate object

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = data[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

```python
#Prepare the data for training
X = pd.DataFrame(np.c_[data['DIS'], data['RM']], columns = ['DIS','RM'])
print(X)
Y = data['MEDV']
print(Y)
```

```python
#randomstate is basically used for reproducing your problem the same every time it is run. 
#If you do not use a randomstate in traintestsplit, every time you make the split you might get a 
#different set of train and test data points
#To understand random_state let us take this small example
from sklearn.model_selection import train_test_split
a , b = np.arange(10).reshape((5,2)),range(5)
#print(a)
#print(b)
print(train_test_split(a,b))
```

```python
from sklearn.model_selection import train_test_split
a , b = np.arange(10).reshape((5,2)),range(5)
#print(a)
#print(b)
print(train_test_split(a,b,random_state = 2))
```
